```python
import torch
from diffusers import StableDiffusionControlNetPipeline
from diffusers.utils import load_image 

#  
image = load_image(
    "https://hf.co/datasets/huggingface/documentation-images/resolve/main/diffusers/input_image_vermeer.png"
)

image
```

OpenCV    .






```python
import cv2
from PIL import Image 
import numpy as np

#  NumPy  . 
image = np.array(image)

# threshold . 
low_threshold = 100
high_threshold = 200

#  . 
image = cv2.Canny(image, low_threshold, high_threshold)
image = image[:, :, None]
image = np.concatenate([image, image, image], axis=2)
canny_image = Image.fromarray(image)  # NumPy  PIL  . 

canny_image
```

      .






```python
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel

canny_controlnet = ControlNetModel.from_pretrained("lllyasviel/sd-controlnet-canny", torch_dtype=torch.float16)
canny_pipe = StableDiffusionControlNetPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", controlnet=canny_controlnet, torch_dtype=torch.float16
)
```

        .






```python
from diffusers import UniPCMultistepScheduler
canny_pipe.scheduler = UniPCMultistepScheduler.from_config(canny_pipe.scheduler.config)
canny_pipe = canny_pipe.to("cuda")

#     seed . 
generator = torch.manual_seed(0)  

#  . 
canny_image_1 = canny_pipe(
    prompt="disco dancer with colorful lights", 
    num_inference_steps=20, 
    generator=generator, 
    image=canny_image
).images[0]

#   . 
canny_image_1.save("/aiffel/aiffel/canny_image.png")  

#   . 
canny_image_1
```

   (negative_propmt)   ,    .

- num_inference_steps:  ,         . default  50.

   ChatGPT   .          .






```python
#  . 
canny_image_2 = canny_pipe(
    prompt="A pretty middle school girl in uniform", 
    num_inference_steps=100, 
    generator=generator, 
    image=canny_image
).images[0]

#   . 
canny_image_2.save("/aiffel/aiffel/canny_image_2.png")  

#   . 
canny_image_2
```

#### Canny

```python
from diffusers.utils import load_image 
from PIL import Image
import cv2
import numpy as np
from diffusers.utils import load_image

# Q.   .
#  . 
canny_image = load_image(
    "https://hf.co/datasets/huggingface/documentation-images/resolve/main/diffusers/input_image_vermeer.png"
)
#canny_image =  np.array(image)

#threshhold . 
low_threshold = 100
high_threshold = 200

#  NumPy  . 
canny_image = np.array(canny_image)

#       . 
zero_start = canny_image.shape[1] // 4
zero_end = zero_start + canny_image.shape[1] // 2
canny_image[:, zero_start:zero_end] = 0

#   NumPy  PIL  . 
canny_image = cv2.Canny(canny_image, low_threshold, high_threshold)
canny_image = canny_image[:, :, None]
canny_image = np.concatenate([canny_image, canny_image, canny_image], axis=2)
canny_image = Image.fromarray(canny_image)  # NumPy  PIL  . 

canny_image
```

```python
from controlnet_aux import OpenposeDetector
from diffusers.utils import load_image 

# Q.    .
#  . 
openpose_image = load_image(
   "https://huggingface.co/datasets/huggingface/documentation-images/resolve/main/diffusers/person.png"
)

# OpenposeDetector    . 
openpose = OpenposeDetector.from_pretrained("lllyasviel/ControlNet")
openpose_image = openpose(openpose_image)

openpose_image
```

Openpose     .






```python
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel 

openpose_controlnet = ControlNetModel.from_pretrained("lllyasviel/sd-controlnet-openpose", torch_dtype=torch.float16)
openpose_pipe = StableDiffusionControlNetPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", controlnet=openpose_controlnet, torch_dtype=torch.float16
)
```

#### ChatGPT   Prompt

Created with AIPRM Prompt "Stable Diffusion Prompt based of your Idea"

Full Detailed Prompt: Subject: A middle school girl Setting: School courtyard during recess Action: Reading a book under a cherry blossom tree Context: Peaceful, tranquil environment, a break from school activities Environment: Blossoming cherry trees, school buildings in the background, other students playing in the distance Lighting: Soft, natural daylight with gentle shadows cast by the cherry blossoms Artist: Inspired by the works of Claude Monet Style: Impressionist Medium: Oil painting Type: Portrait Color Scheme: Pastel colors, dominated by pink and green hues Computer Graphics: High resolution, digital oil painting simulation Quality: Ultra-detailed, emphasizing texture and light

Positive Prompt: middle school girl, school courtyard, reading under cherry blossom, peaceful scene, cherry trees, distant students, natural daylight, soft shadows, Monet inspiration, Impressionist style, oil painting, portrait, pastel pink and green, digital high resolution, ultra-detailed texture and light

Negative Prompts: no cityscape, no indoor setting, no night scene, no rain, no winter, no abstract, no modernism, no acrylic, no sculpture, no monochrome, no low resolution, no harsh lighting, no overcrowded scene

```python
from diffusers import StableDiffusionControlNetPipeline, ControlNetModel, UniPCMultistepScheduler  

# Q.   .
# Edge Detection Openpose, 2  controlnets  . 
controlnets = [canny_controlnet, openpose_controlnet] 

#  controlnets  . 
pipe = StableDiffusionControlNetPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5", controlnet=controlnets, torch_dtype=torch.float16
)
pipe.scheduler = UniPCMultistepScheduler.from_config(pipe.scheduler.config)
pipe = pipe.to("cuda")

#  . 
prompt =  'middle school girl, school courtyard, reading under cherry blossom, peaceful scene, cherry trees, distant students, natural daylight, soft shadows, Monet inspiration, Impressionist style, oil painting, portrait, pastel pink and green, digital high resolution, ultra-detailed texture and light' # [[YOUR CODE]]
negative_prompt =  'no cityscape, no indoor setting, no night scene, no rain, no winter, no abstract, no modernism, no acrylic, no sculpture, no monochrome, no low resolution, no harsh lighting, no overcrowded scene' # [[YOUR CODE]]

# seed . 
generator = torch.manual_seed(2023)

images = [canny_image, openpose_image]

#  . 
image = pipe(
    prompt= [prompt, negative_prompt],
    #negative_prompt = negative_prompt,
    num_inference_steps=100, 
    generator=generator, 
    image=images
).images[0]

#   .
image.save("/aiffel/aiffel/multi_controlnet_output.png")

#   .  
image
```
