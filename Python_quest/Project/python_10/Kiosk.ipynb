```python
# 문제 3-3. 클래스 업데이트  
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요! 

class Kiosk:
    def __init__(self):
        self.menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
        self.price = [2000, 3000, 3000, 2500, 2500, 3000]

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ":", self.price[i])

    # 주문 메서드
    def menu_select(self):
        self.order_menu = []
        self.order_price = []
        
        n = 0
        #self.price_sum = 0
        
        while n < 1 or n > len(self.menu):
            # 음료 번호 입력
            n = int(input("\n음료 번호를 입력하세요 : "))
            
        # HOT/ICE 여부 입력
        t = ""
        while t not in ("1", "2"):
            t = input("\nHOT 음료는 1을, ICE 음료는 2를 입력하세요 : ")
            if t == "1":
                self.temp = "HOT"
            elif t == "2":
                self.temp = "ICE"

        # 주문받은 음료명과 가격을 주문 메뉴/가격 리스트에 추가한다.
        self.order_menu.append(self.temp + " " + self.menu[n - 1])
        self.order_price.append(self.price[n - 1])
        self.price_sum = self.price[n - 1]

        # 메뉴와 가격 출력
        print(self.temp, self.menu[n - 1], ":", self.price[n - 1])

        # 추가 주문
        while n != 0:
            n = int(input("\n추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            
            if 1 <= n <= len(self.menu):
                t = 0
                while t not in (1, 2):
                    t = int(input("\nHOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"

                self.order_menu.append(self.temp + " " + self.menu[n - 1])
                self.order_price.append(self.price[n - 1])
                self.price_sum += self.price[n - 1]

                print("\n추가 주문 음료", self.temp, self.menu[n - 1], ":", self.price[n - 1])
                print("합계 :", self.price_sum, "원")
            else:
                print("없는 메뉴 번호입니다.")
                
                

        print("\n주문이 완료되었습니다.\n", self.order_menu, self.order_price)

    # 지불
    def pay(self):
        print(f"총 주문 금액은 {self.price_sum}입니다.")
        payment = 0
        while payment not in (1, 2):
            payment = int(input("\n지불 수단을 선택해 주세요.\n현금: 1, 카드: 2 :"))
            if payment == 1:
                print("직원을 호출하겠습니다.\n")
            elif payment == 2:
                print("IC카드를 방향에 맞게 삽입해 주세요.\n")

    # 주문서 출력 
    def table(self):
        print('+' + '-' * 30 + '+')
        for i in range(5):
            print('|' + ' ' * 30 + '|')
            
        for i in range(len(self.order_menu)):
            print(self.order_menu[i], ":", self.order_price[i])
            
        print("합계 금액 :", self.price_sum) # sum(order_price))

        for i in range(5):
            print('|' + ' ' * 30 + '|')
        print('+' + '-' * 30 + '+')
```

```python
k = Kiosk()
k.menu_print()
k.menu_select()
k.pay()
k.table()
```
