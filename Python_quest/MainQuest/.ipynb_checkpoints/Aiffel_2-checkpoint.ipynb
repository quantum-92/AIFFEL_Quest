```python
import pandas as pd

# 이름을 '키'라 가정. 즉, 이름이 동일한 사람이 없다.
class Aiffel():

    #grews = pd.DataFrame()
    
    def __init__(self):
        self.student_count = 0
        self.grews = pd.DataFrame()
    
#         self.grews = {}
        
#         self.net_score = 0
#         self.total_score = 0
#         self.total_penalty = 0
        #self.name = name
        #self.student_id = self.get_student_id(enroll_year, batch, course)

#     def get_student_id(enroll_year, batch, course):

    # 딕셔너리를 입력 받아 학번 생성
    def add_grews(self, grew_dic):
        # 학번 생성

        # 초기화
        self.student_count = 0
        
        # 이름 순서대로 학번 생성    
        for name in sorted(grew_dic):
            self.student_count += 1    # 학생수 1 증가
            grew_dic[name] = f"23252{self.student_count:02d}"  # 코스 정보가 없으므로 모드 코어라 가정
        
        self.grews = pd.DataFrame({
            'name':list(grew_dic.keys()),
            'id':list(grew_dic.values())
        })
        course_chr = list(x[-3] for x in self.grews['id'])
        self.grews['course'] = ['리서치' if x == '1' else '코어' for x in course_chr]
        self.grews['net_score'] = 0
        self.grews['total_score'] = 0
        self.grews['total_panalty'] = 0
        
    def get_id_num(self):
        return(self.student_count)
        
    def score(self, name, quest_socre):
        if quest_socre < 1 or quest_socre > 3:
            print()
            return
    
        self.grews.loc[self.grews['name'] == name, 'total_score'] += quest_socre    # 주어진 score의 누적
        self.grews.loc[self.grews['name'] == name, 'net_score'] += quest_socre      # 

    # 패널티도 1 ~ 3점?
    def panalty(self, name, quest_panalty):
        if quest_panalty < 1 or quest_panalty > 3:
            print()
            return

        self.grews.loc[self.grews['name'] == name, 'total_panalty'] += quest_panalty    # 주어진 score의 누적
        self.grews.loc[self.grews['name'] == name, 'net_score'] -= quest_panalty      # 
    
    def display_info(self):
        self.grews.to_csv('grews.csv', index = False)
            
    def del_grew(self, name):
        self.grews = self.grews[self.grews['name'] != name]
        
    
```

```python
grews = {"김건": 0, "김경민": 0, "김소연": 0, "김영신": 0, "박근수": 0, \
         "박영수": 0, "박영준": 0, "서원영": 0, "신지만": 0, "심지안": 0, \
         "이규상": 0, "이윤상": 0, "이진영": 0, "임현석": 0, "전요한": 0, \
         "전재영": 0, "조대희":0}

aiffel = Aiffel()
aiffel.add_grews(grews)
```

```python
import random as r

class Aiffel_Guild(Aiffel):
    def __init__(self):
        super().__init__()
        self.guild_score_data = 0
        self.guild_dict = {}
    
    def map_guild(self):
        guild_list = ('힌튼', '알트만', '응', '하사비스')
        
        for name in self.grews['name']:
            self.guild_dict[name] = guild_list[r.randint(0, 3)]
        
        
    def guild_score(self, guild, quest_score):
        if (-3 <= quest_score <= 3 and quest_score != 0):
            self.guild_score_data += quest_score
            
            for grew_name, grew_guild in self.guild_dict.items():
                if grew_guild == guild:
                    if (quest_score > 0):
                        super().score(grew_name, quest_score)
                    else:
                        super().panalty(grew_name, - quest_score)
        
```
