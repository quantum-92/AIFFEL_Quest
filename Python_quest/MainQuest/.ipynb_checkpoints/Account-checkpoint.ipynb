```python
import random as r

#계좌 클래스
class Account():
    bank = "SC은행"       # 은행명
    total_accounts = 0   # 총계좌수
    
    # 계좌생성
    def __init__(self, name, balance):
        # 계좌번호를 111-11-111111 모양으로 랜덤 생성
        self.account_number = f"{r.randint(0, 999):03d}-{r.randint(0,99):02d}-{r.randint(0,999999):06d}"
        self.name = name               # 고객명
        self.balance = balance         # 잔고
        self.deposit_count = 0         # 입금횟수
        self.deposit_list = []         # 입금 히스토리
        self.withdraw_list = []        # 출금 히스토리
        Account.total_accounts += 1    # 총계좌수 1증가
        print(f"{Account.bank}에 {self.name}님의 계좌를 생성하였습니다.\n 계좌번호: {self.account_number}\n 잔고: {self.balance}\n")
        
    # 총계좌수 조회
    def get_account_num(self):
        print(f"{Account.bank}의 총 계좌수는 {Account.total_accounts}개입니다.\n") # 클래스변수를 조회하여 출력한다.

    # 입금
    def deposit(self, amount):
        print(f"{self.name}님의 {self.account_number} 계좌에 {amount:,}원을 입금하겠습니다.") # 입금을 시도한다는 메시지 출력
        if (amount < 1):         # 입금액이 1원 이상인지 확인
            print(" 입금 실패했습니다.1원 이상만 입금가능 합니다.\n")
        else:
            interest = 0
            self.deposit_count += 1
            if self.deposit_count <= 5:
                interest = int(self.balance * 0.01)
                self.balance += interest
                print(f"이자가 {interest:,}원 지급되었습니다.") 
            self.balance += amount              # 잔고 += 입급액
            self.deposit_list.append(f"입금액: {amount:,}, 이자: {interest:,}, 잔고: {self.balance:,}")

            print(f" 입금 성공했습니다. 잔고는 {self.balance:,}원입니다.\n")
    
    # 출금
    def withdraw(self, amount):
        print(f"{self.name}님의 {self.account_number} 계좌에서 {amount:,}원을 출금하겠습니다.") # 출금을 시도한다는 메시지 출력
        if (amount > self.balance):   # 잔고가 출금액 이상인지 확인
            print(f" 출금 실패했습니다. 잔고가 부족합니다.(잔고:{self.balance:,})\n")
        else:
            self.balance -= amount    # 잔고 -= 출금액
            self.withdraw_list.append(f"출금액: {amount:,}, 잔고: {self.balance:,}")
                                      
            print(f" 출금 성공했습니다. 잔고는 {self.balance:,}원입니다.\n")
    
    # 계좌정보 출력
    def display_info(self):
        print(f"은행이름: {Account.bank}, 예금주: {self.name}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")
    
    # 입금 히스토리 출력
    def deposit_history(self):
        print("입금 내역:")
        for h in self.deposit_list:
            print(h)
        print()
    
    # 출금 히스토리 출력
    def withdraw_history(self):
        print("출금 내역:")
        for h in self.withdraw_list:
            print(h)
        print()
        
```

```python
accounts = [Account("이혁희", 5000000), \
    Account("하사비스", 300000), \
    Account("알트만", 4000000)]

print("잔고 100만원 넘는 계좌:")
for a in accounts:
    if a.balance > 1000000:
        a.display_info()
```

```python
# 계좌 생성
a = Account("이혁희", 3000000)

# 입출금 반복. 
# 5회까지 이자가 지급되는 것을 확인하기 위하여 입금 6회 반복
a.deposit(500000)
a.withdraw(1000000)
a.deposit(100000)
a.deposit(100000)
a.deposit(100000)
a.deposit(100000)
a.deposit(100000)
a.withdraw(500000)

# 입출금 히스토리 출력
a.deposit_history()
a.withdraw_history()
```
